# Boolean Mask Arrays

```python
import numpy as np
my_array = np.array([-17, -4, 0, 2, 21, 37, 105]) 
my_array
```

```python
zero_mod_7_mask=(my_array % 7 == 0)
zero_mod_7_mask
```

```python
sub_array=my_array[zero_mod_7_mask]
sub_array
```

```python
#take care that if you tried by 1/0 instead of True/False, it's not boolean masking at all 
list_2 = [0,0,1,0,1,0,1]
not_zero_mod_7_mask= np.array(list_2)
my_array[not_zero_mod_7_mask]                     #show index 1,0
```

```python
my_array[my_array>0]
```

```python
mod_test=(0==my_array%7)
mod_test
```

```python
poistivee=my_array>0
poistivee
```

```python
combine_mask=np.logical_and(poistivee,mod_test)
combine_mask
```

```python
poistivee&mod_test
```

```python
my_array[combine_mask]
```
