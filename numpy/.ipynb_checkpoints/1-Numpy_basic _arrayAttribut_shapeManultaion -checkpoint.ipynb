# Create NumPy arrays using Python's "array like" data types


```python
my_first_array = np.arange(11)    
print (my_first_array)
```

```python
my_list = [-17, 0, 4, 5.0, 9]             # diffrent datatype
my_array_from_list = np.array(my_list)   #array is func  -->  same data type  
my_array_from_list                        # float 
```

```python
my_tuple =(14, -3.54, 5+7j)                # diffrent datatype 
my_array_from_tuple = np.array(my_tuple)   #same datatype   complex data type   with zero  
my_array_from_tuple
```

```python
my_array_from_tuple.dtype        # to show  item type
```

# linspace(), zeros(), ones(), and NumPy data types


```python

import numpy as np
```

```python
my_linspace=np.linspace(5, 15, 9)   # (st, ed  is included   , size  )
my_linspace                           # internal calc step , step =1.25
```

```python
my_linspace=np.linspace(5, 15, 9, retstep=True)   # (st, ed  is included   , size  ) ,retstep for  return step 
my_linspace
```

```python
type(my_linspace)    #  type  is tuple  of  (array , step )
```

```python
my_linspace[1]   #retuen step 
```

# Array Shape Manipulation


```python
import numpy as np
```

# reshape
from: https://numpy.org/doc/stable/reference/generated/numpy.reshape.html#numpy.reshape  
NOTE: This will be a new view object if possible; otherwise, it will be a copy.


```python
my_start_array = np.arange(24)
my_start_array
```

```python
my_3d_array=my_start_array.reshape((2,3,4))     # reshape  return copy of array
my_3d_array
```

```python
my_3d_array[1,0,1]=123
my_3d_array[1,0,2]=555
my_3d_array
```

# ravel function 

```python
my_ravel_array=my_3d_array.ravel()        # ravel return flat array /1d array
my_ravel_array
```

```python
my_ravel_array.shape
```

# squeeze function 
remove any axis that have vlaue 1

```python
my_3d_array
```

```python
my_3d_array.shape
```

```python
my_3d_array.squeeze()  # no change   ,because all axis not equal 1
```

```python
arr=np.arange(24).reshape((1,6,4))
arr
```

```python
arr.shape
```

```python
arr=arr.squeeze()
arr
```

```python
arr.shape
```

```python
arr=np.arange(30).reshape((6,1,5))
arr
```

```python
arr=arr.squeeze()
arr
```

```python
arr.shape
```

```python
arr=np.arange(24).reshape((1,24,1))
arr
```

```python
arr=arr.squeeze()
arr                            # 1d  
```

```python
#  if need squeeze  only first  Dima
arr=np.arange(24).reshape((1,24,1))
arr=arr.squeeze(axis=0)              # axis =0 mean only change first  dim if have value 1 
arr
```

```python
#  if need squeeze  only last  Dim
arr=np.arange(24).reshape((1,24,1))
arr=arr.squeeze(axis=2)              # axis =2 mean only change last  dim if have value 1 
arr
```

```python
#  if need squeeze  only first  Dima
arr=np.arange(24).reshape((1,24,1))
arr=arr.squeeze(axis=1)              # error 
arr
```
