#### 64. Consider a given vector, how to add 1 to each element indexed by a second vector (be careful with repeated indices)? (★★★)

````python
import numpy as np

a=np.zeros(10)
idx=np.array([1,3,4,1])
a[idx]+=1
print(a)
"""
That’s because NumPy’s fancy indexing does not accumulate on repeats —
it applies the operation on a temporary copy, then writes back once.
"""
# correct sol
print()
a=np.zeros(10)
np.add.at(a, idx, 1)
print(a)

# other solution
a=np.zeros(10)
frq=np.bincount(idx,minlength=10)
a+=frq
print(a)
````

#### 85. Create a 2D array subclass such that Z[i,j] == Z[j,i] (★★★)

````python
import numpy as np

class Symetric(np.ndarray):
    def __setitem__(self, index, value):
        i,j = index
        super(Symetric, self).__setitem__((i,j), value)
        super(Symetric, self).__setitem__((j,i), value)

Z = Symetric((5, 5))
Z[:,:]=1
Z[1, 3] = 7

print(Z)
print("Z[3,1] =", Z[3,1])   # automatically symmetric
print("Z[1,3] =", Z[1,3])
````

#### 89. How to get the n largest values of an array (★★★)

````python
Z = np.arange(10000)
np.random.shuffle(Z)
Z.sort()
print(Z[-5::])

###############################
Z = np.arange(10000)
np.random.shuffle(Z)
n=5
Z[Z.argsort()[-n::]]


############################
print (Z[np.argpartition(-Z,n)[:n]])
````

🔹 What argpartition does

Think of it like the quickselect algorithm (used in finding the k-th largest element).

It partitions the array around a pivot so that:

All elements before position k are smaller (not sorted).

All elements after position k are larger (not sorted).

It doesn’t fully sort → just ensures the relative position of the "cut".

That’s why it’s O(N) instead of O(N log N).   
</br>




🔹 Visual intuition

Suppose array = [10, 3, 15, 7, 8, 23, 1], and we want top n=3.

argpartition(Z, -3) rearranges indices so that the last 3 slots contain the 3 largest numbers.

But within those slots, the numbers can appear in any order.

If you want them ranked → you sort just that slice.

#### 90. Given an arbitrary number of vectors, build the cartesian product (every combination of every item) (★★★)

````python
def cartesian(arrays):
    arrays = [np.asarray(a) for a in arrays]
    shape = (len(x) for x in arrays)

    ix = np.indices(shape, dtype=int)
    ix = ix.reshape(len(arrays), -1).T
    print(ix)
    for n, arr in enumerate(arrays):
        ix[:, n] = arrays[n][ix[:, n]]  # replace  col  in idx by colum in array number  coresspond to

    return ix

print (cartesian(([1, 2, 3], [4, 5], [6, 7])))
````

#### 97. Considering 2 vectors A & B, write the einsum equivalent of inner, outer, sum, and mul function (★★★)

````python
A=np.array([1,2,3])
B=np.array([1,1,1])

print(np.einsum('i,i->',A,B)   )# inner or dot product
print(np.einsum('i,j->ij',A,B)  ) # out
print(np.einsum('i,i->i',A,B)  )  #element wise multi
print(np.einsum('i->',A)  )  #element wise multi


````

Think of einsum as: "multiply arrays where indices match, then sum over any index not in the output specification."

So: np.einsum = "multiply over matching indices, sum over repeated ones, keep the rest in the output order."
