### Copy your single variable linear regression function you created in practical work 1

```python
def  GD_func(x,y,alpha,max_iter,grad_check,loss_check):
    th_0 = 0.0
    th_1 = 0.0
    m=len(x)
    loss=[]
    th_0_array=[]
    th_1_array=[]
    # Gradient descent loop
    for epoch in range(max_iter) :
        print(f"****************** Iteration {epoch}********************")
        # Calculate predictions
        h =th_0 + th_1 * x 
        print(f"h(x)= \n{h}")
        e = h - y
        print(f"error vector=\n {e}\n")
        # Compute gradients
        j =e@e/(2*m) #(1 / (2 * m)) * np.dot(error, error)  # Dot product approach
        th_0_array.append(th_0)
        th_1_array.append(th_1)
        loss.append(j)
        print(f"j ={j} ")
        d_th_0 = e.sum()/m              #(1 / (m)) * np.sum(y_pred - y)*1
        d_th_1 = (e@x)/m               #(1 / (m)) * np.dot(y_pred - y,x)
        print(f"gradaint = \n[ {d_th_0} \n {d_th_1} ]\n")
        # cond check 
        grad=np.array([[d_th_0],[d_th_1]])
        gradaint_norm = np.linalg.norm(grad)
        if gradaint_norm<grad_check:
            print("****************** Training Report ********************\n")
            print(f" Gradient Descent converged after {epoch+1} iterations\n")
            print(f" theta_0_Opt = {th_0}")
            print(f" theta_1_Opt= {th_1}\n")
            print(f" Error Vector:\n{e}\n")
            print(f" cost {j}\n")
            break
        if epoch>0 and abs(loss[epoch-1]-loss[epoch])<loss_check:
            print("****************** Training Report ********************\n")
            print(f" Gradient Descent converged after {epoch+1} iterations\n")
            print(f" theta_0_Opt = {th_0}")
            print(f" theta_1_Opt= {th_1}\n")
            print(f" Error Vector:\n{e}\n")
            print(f" cost {j}\n")
            break
        # Update parameters (simultaneously)
        th_0-=alpha*(d_th_0)
        th_1-=alpha*(d_th_1)
   
    print(f" h(x) = y_predict: \n{h}\n")
    print(f" y_actual :\n{y}\n")
    return th_0_array,th_1_array,loss




```

```python
th_0_array,th_1_array,loss=GD_func(X,y,.001,500,.001,.001)
```

## Stochastic GD for Single Variable LR Implementation
### Step 1: Shuffle your data

#### Concatenate X,y data in one array using np.concatenate()

```python
X=X.reshape(-1,1)
y=y.reshape(-1,1)

data=np.concatenate(   (X,y) ,axis=1 )
print(data.shape)
data
```

#### Shuffle your data using np.random.shuffle()
#### Set random seed to 101 to obtain the same result as the one we do here

```python
# call  above cell before this

np.random.seed(101)
np.random.shuffle(data)  
data
```

#### Reassign first column to variable X and second column to variable y.

```python
X=data[:,0]
Y=data[:,1]
print(X ,"\n",Y)
```

### Step 2: Parameters initialization (theta0 = 0, theta1 = 0, alpha = 0.001)
- In this step we will also get the number of data points <b>(m)</b>.
- We will also define empty lists to save losses, thetas in all iterations.

```python
th0=th1=0
alpha=.001
m=data.shape[0]
```

### Step 3: Claculate the prediction for the first data point

```python
h_0=X[0]*th1+th0
h_0
```

### Step 4: Claculate the error for the first data point prediction h(0) - y(0)

```python
e_0=h_0-Y[0]
e_0
```

### Step 5: Claculate the cost for the first data point prediction $\mathbf{\frac{e^2}{2}}$

```python
j_0=(e_0*e_0)/2
j_0
```

### Step 6: Claculate the gradient components for the first data point 
#### d_th0 = e
#### d_th1 = e * X[0]

```python
d_th0=e_0
d_th1=e_0*X[0]
print(d_th0,"\n",d_th1)
```

### Step 7: Update theta values

```python
th0-=alpha*d_th0
th1-=alpha*d_th1
print(th0,"\n",th1)
```

### Step 8: Iterate through all data points to finish the first epoch
#### Save thetas and loss each iteration i.e. with each update.
#### Reinitialize parameters in the begining: th0 = 0, th1 = 0.
#### Plot the learning curves to explore parameters update in single epoch.

```python
alpha=.001
th0 = 0.0
th1 = 0.0
m=len(X)
loss=[]
th0_array=[]
th1_array=[]
# Gradient descent loop
for i in range(m) :
    h =th1 * X[i]+th0
    e = h - Y[i]
    # Compute gradients
    j =(e*e)/2                                    #(1 / (2 * m)) * np.dot(error, error)  # Dot product approach
    th0_array.append(th0)
    th1_array.append(th1)
    loss.append(j)
    d_th0 = e                             #(1 / (m)) * np.sum(y_pred - y)*1
    d_th1 = (e*X[i])                     #(1 / (m)) * np.dot(y_pred - y,x)
    # Update parameters (simultaneously)
    th0-=(alpha*d_th0)
    th1-=(alpha*d_th1)




 
```

### Step 9: Make outer loop to perform as many epochs as you need to reach the stop conditions
### Remember the following:

##### Very Important notes:
- <font color='red'><b>Data should be shuffled for stochastic GD to work fine.</b></font>
- <font color='red'><b>When you shuffle the data, input varibles and equivelant target label should be together. i.e. do not shuffle X and y separately or your data will be missed up.</b></font>
- <font color='red'><b>Save your losses and thetas for each iterations (not each epoch).</b></font>
- <font color='red'><b>Plot loss vs. iterations (not epochs). This will allow you to see the difference between SGD (stochastic GD) and batch GD.</font>
- <font color='red'><b>Convergence check should be performed each epoch not each iterations. This allows you to perform more accurate decision. This is because the noisy process could lead to a deciving decision.</b></font>

##### The following results uses the hyper-parameters alpha = 0.001, max. epochs = 500, and 0.001 for both stop conditions.

```python

alpha=.001
th0 = 0.0
th1 = 0.0
m=len(X)
loss=[]
th0_array=[]
th1_array=[]
epochs=500
for epoch in range(epochs):
    print(f"****************** Epoch {epoch} ********************")
    # Gradient descent loop
    for i in range(m) :
        h =th1 * X[i]+th0
        e = h - Y[i]
        # Compute gradients
        j =(e*e)/2                                    #(1 / (2 * m)) * np.dot(error, error)  # Dot product approach
        th0_array.append(th0)
        th1_array.append(th1)
        loss.append(j)
        d_th0 = e                             #(1 / (m)) * np.sum(y_pred - y)*1
        d_th1 = (e*X[i])                     #(1 / (m)) * np.dot(y_pred - y,x)
        # Update parameters (simultaneously)
        th0-=(alpha*d_th0)
        th1-=(alpha*d_th1)
        
    print(f" cost = {j}\n") 
    grad=np.array([[d_th0],[d_th1]])
    grad_norm=np.linalg.norm(grad)
    print(f" grad vect= =\n {grad}\n") 
    print(f" grad  norm={grad_norm}\n") 
    print(f" new_theat0= {th0}\n") 
    print(f" new_theat1 ={th1}\n")
    if( grad_norm<.001):
        print(f"****************** Training Report ********************")
        print(f" opt_theat0= {th0_array}\n") 
        print(f" opt_theat1 ={th1_array}\n")
        print(f" cost = {j}\n") 
        break
    if( epoch>0 and np.absolute(loss[epoch*m-1]-loss[-1])<.001):
        print(f"****************** Training Report ********************")
        print(f" Gradient Descent converged after {epoch+1} epochs\n") 
        print(f" opt_theat0= {th0_array[-1]}\n") 
        print(f" opt_theat1 ={th1_array[-1]}\n")
        print(f" cost = {j}\n") 
        break
   




 
```

### Make your implementation as a function

```python
def GD_statistic(X,Y,epochs=500,alpha=.001,grad_check=.001 ,conver_check=.001):
        
    th0 =th1= 0.0
    m=len(X)
    loss=[]
    th0_array=[]
    th1_array=[]
    for epoch in range(epochs):
        print(f"****************** Epoch {epoch} ********************")
        # Gradient descent loop
        for i in range(m) :
            h =th1 * X[i]+th0
            e = h - Y[i]
            # Compute gradients
            j =(e*e)/2                                    #(1 / (2 * m)) * np.dot(error, error)  # Dot product approach
            th0_array.append(th0)
            th1_array.append(th1)
            loss.append(j)
            d_th0 = e                             #(1 / (m)) * np.sum(y_pred - y)*1
            d_th1 = (e*X[i])                     #(1 / (m)) * np.dot(y_pred - y,x)
            # Update parameters (simultaneously)
            th0-=(alpha*d_th0)
            th1-=(alpha*d_th1)
        print(f"cost ={j}")
        grad=np.array([[d_th0],[d_th1]])
        grad_norm=np.linalg.norm(grad)
        print(f" grad vect= =\n {grad}\n") 
        print(f" grad  norm={grad_norm}\n") 
        print(f" new_theat0= {th0}\n") 
        print(f" new_theat1 ={th1}\n")
        if( grad_norm<grad_check):
            print(f"****************** Training Report ********************")
            print(f" opt_theat0= {th0_array}\n") 
            print(f" opt_theat1 ={th1_array}\n")
            print(f" cost = {j}\n") 
            break
        if( epoch>0 and np.absolute(loss[-(m+1)]-loss[-1])<conver_check):   # -(m+1)= epoch*m-1
            print(f"****************** Training Report ********************")
            print(f" Gradient Descent converged after {epoch+1} epochs\n") 
            print(f" opt_theat0= {th0_array[-1]}\n") 
            print(f" opt_theat1 ={th1_array[-1]}\n")
            print(f" cost = {j}\n") 
            break
        
    return  th0_array,th1_array,loss 




 
```
