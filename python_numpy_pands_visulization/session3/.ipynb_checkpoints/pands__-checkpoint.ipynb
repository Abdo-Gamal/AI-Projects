## can make series from list , numpy array , or dictionary 



```python
lablels=['a','b','c','d']
my_data=[11,22,32,42]
arr=np.array(my_data)
d={'a':12,'b':34,'c':44}
```

```python
###############using list      and array 
# default index
s=pd.Series(my_data)
s1=pd.Series(arr)
# custom index 
s2=pd.Series(my_data, lablels)
print(s,s1,s2)

###############using dic       and array 

s3=pd.Series(d)   # index is key & value is 
s3
```

# Know the dublicated 

```python
data ={'company':['goog','goog','goog','msf','msf','fb','fb'],
        'person':['sam','sam','charlie','amy','vanessa','carl','sarah'],
       'sales':[200,200,120,340,124,243,350]
      
      }

df = pd.DataFrame(data)
print(df)                            

df[df.duplicated()] #    
```

```python
# remove dublicate 
df.drop_duplicates( inplace= True)   # must be put  inplace 
```
