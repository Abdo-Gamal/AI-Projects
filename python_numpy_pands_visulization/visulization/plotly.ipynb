```python
import pandas as pd    
import plotly.express as px 
```

```python
df =px.data.gapminder()
df.head()
```

```python
px.scatter(df,y='lifeExp',x='gdpPercap')
# can zoom in ,zoom out ,donlowd it 
```

```python
px.scatter(df,y='lifeExp',x='gdpPercap',size='pop') # make size pased on population
```

```python
px.scatter(df,y='lifeExp',x='gdpPercap',size='pop',size_max=60) # maxsumixe point sixe 
```

```python
px.scatter(df,y='lifeExp',x='gdpPercap',size='pop',size_max=60,color='continent') # make color
```

```python
px.scatter(df[df['year']==2007],y='lifeExp',x='gdpPercap',size='pop',size_max=60,color='continent') #choose one year to plot
```

```python
px.scatter(df[df['year']==2007],y='lifeExp',x='gdpPercap',size='pop',size_max=60,color='continent',hover_name='country') #choose hover_future
```

```python
px.scatter(df[df['year']==2007],y='lifeExp',x='gdpPercap',size='pop',size_max=60,color='continent',hover_name='country',title='abdo gamal') #add title
```

```python
px.scatter(df,y='lifeExp',x='gdpPercap',size='pop',size_max=60,color='continent',hover_name='country',title='abdo gamal',animation_frame='year',height=1000) #
```
