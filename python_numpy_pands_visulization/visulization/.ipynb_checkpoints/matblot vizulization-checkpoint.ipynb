```python
import pandas as pd # library for data manipulation
import matplotlib.pyplot as plt # library for visualization
import seaborn as sns # library for visualization
```

```python
df_cars=pd.read_csv(r"bar_chart_data.csv")
df_cars
```

```python
#        bar 
plt.figure(figsize=(10,10))
plt.bar(x=df_cars['Brand'],height=df_cars['Cars Listings'])


plt.show()
```

```python
#    
#        bar 
plt.figure(figsize=(10,10))
plt.bar(x=df_cars['Brand'],height=df_cars['Cars Listings'])

plt.title('tttttttttttt',fontsize=22)
plt.ylabel('no' ,fontsize=22)
plt.xlabel('car',fontsize=33)


plt.show()
```

```python
df_engine_fuel=pd.read_csv('pie_chart_data.csv')
df_engine_fuel
```

```python
#          par  or pie  

plt.figure(figsize=(5,5))
plt.pie(df_engine_fuel['Number of Cars'],labels=df_engine_fuel['Engine Fuel Type'].unique())
plt.title('dssdsdds',)
plt.show()
#pie take (data , labels )
```

```python
df=pd.read_csv('stacked_area_chart_data.csv')
df
```

```python
#          line   or  area    
#       area 

plt.figure()
             # X axis     Y axis   y asix ..   
plt.stackplot( df['Year'],df['Gas'],df['Petrol'],df['Diesel'] ,
                  colors=['red','blue','yellow'])# take x, many y
plt.title('ddssd')
plt.ylabel('area')
plt.xlabel('price')
plt.legend(labels=['P','G','F'])   


plt.show()
```

```python
df=pd.read_csv('line_chart_data.csv')
df.head()
```

```python
#     line  or area   
#    date  
#     
#     object
#   
df.info()
```

```python
df['newdate']=pd.to_datetime(df['Date'])
df.info()
```

```python
#         
df['month']=df['newdate'].apply(lambda x:x.month)
df['year']=df['newdate'].apply(lambda x:x.year)
df['day']=df['newdate'].apply(lambda x:x.day)
df.info()
df.head()
```

```python
plt.figure(figsize=(20,10))

plt.plot(df['newdate'],df['GSPC500'])#       
plt.plot(df['newdate'],df['FTSE100']) #      

plt.show()
```

```python
df=pd.read_csv('histogram_data.csv')
df.head()

#   distribution 
# price
```

```python
plt.figure()
plt.hist(df['Price'])
plt.show()
#               
# mode =200
#    200  250   
```

```python
#            price  ,area 
# use  scatter plot  

plt.figure()
       #              
       #               
plt.scatter(df['Price'],df['Area (ft.)'],c=df['Building Type'])

plt.show()
```
