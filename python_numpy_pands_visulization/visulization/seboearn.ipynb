```python
import seaborn as sns
# من مميزات المكتبه دى انها بتختصر عليك شغل زى اسم العموض واسم الصف   
#والtitle  
#
```

```python
tips=sns.load_dataset('tips')  # بقشيش
#  اقدر احسب  مين اكتر كرم الستات ولا الرجاله  
# المدخنين  والا اللى مش بيشربه هم اللى بيدفعه اكتر 
# انهى يوم اللى بندفع فيه اكتر 
#انهى وجبه بندفع فيها  بثشيش اكتر  
```

```python
tips.head()
```

```python
sns.histplot(data=tips,x='sex')
```

```python
sns.histplot(data=tips,x='total_bill',)  # الناس بتدفع من 10 ل 20 دولار 
```

```python
sns.histplot(data=tips,x='total_bill',kde=True)
```

```python
sns.histplot(data=tips,x='total_bill',kde=True,hue='sex')  # heu بيعمل جروب على اساس حاجه معينه 
```

```python
#show all relation  bewean future
sns.pairplot(tips)

# هتلافى ان فى عهلاقه قويه بين البقيش والفاتوره 
```

```python
# هيرسم نفس الجراف اللى فوق بس اللرجاله لوحده والستات لوحدهم  
sns.pairplot(tips,hue='sex')
```

```python
sns.boxplot(data=tips,x='total_bill')

# outlier  over 40 
```

```python
sns.boxplot(data=tips,x='total_bill',y='time')
#هقدر ارد على سوال بيدفعه فلوس اكتر امتى 
#طبعا فى العشاء
```

```python
sns.boxplot(data=tips,x='total_bill',y='time',hue='smoker')

# الناس اللى بدخن بتدفع اكتر فى العشاء  
# والعكس فى الفطار 
```

```python
#  لو عندى    categoral date   
# هرسم   bar 
sns.barplot(data=tips, x='sex', y='total_bill')
```

```python
# count plot 

sns.countplot(data=tips,x='sex')

#هتلاحظ ان  عدد الرجاله اكتر من الستات  فكده  الاستنتاج اللى فات ان الرجاله اكرم من الستات ممكن مش صح 
```
