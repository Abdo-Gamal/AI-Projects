## type of error

```python
#ex on exception
def division(nu1,nu2):
    return nu1/nu2

# exception like throw storn in your face  you shoud handle it 
res=division(2,0)
```

```python
#solve Exception

def division(nu1,nu2):
    try:
      return nu1/nu2
    except:
      print("zero division error ")

# exception like throw storn in your face  you shoud handle it 
res=division(2,0)
```

```python
# else and final 

def division(nu1,nu2):
    try:
      return nu1/nu2
    except:
      print("zero division error ")
    else: 
        print("done")    # enter  in else if body of try excute use to close file
    finally:               
        print('final')   # alwas enter in final 


res=division(2,0)
print('\n')
res=division(2,2)
```

```python
# Copy all the content of one file to another file in uppercase
# capitalize func make first lwtter  captal 
# To open the first file in read mode
f1=open('file.txt','r')
# To open the second file in write mode
f2=open('file2.txt','w')

for line in f1:
    f2.write(line.upper())

f1.close()
f2.close()
```
