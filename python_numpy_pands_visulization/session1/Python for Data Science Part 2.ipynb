# filter 

```python
def check_even(x):
    return x%2==0

check_even(5)
L=[1,2,3,4,5,6,7,99]
L_filter=[]
for i in L:
    if check_even(i): 
        L_filter.append(i)

print(L_filter)
## اقدر اختصر الكود لكده 

list(filter(check_even,L))


```

# map




### map convert thing to other 
### filter return ture or flase and return some elem that complete the condition 
### map make process on elem and return it after process 


```python
def square(x):
    return x**2
L=[1,2,3,4,5,6,7,99]

list(map(square,L))# every item become  x**2  
```

## time 

```python
import datetime as dt
t= dt.time(4,20,1)


print(t)
print('hour : ',t.hour)
print('minute : ',t.minute)
print('second : ',t.second)
print('microsecond : ',t.microsecond)



```

## date 

```python
today=dt.date.today()
print(today)
print('Ctime :',today.ctime())
print('tuple :',today.timetuple())
print('ordinal :',today.toordinal())

print('year :',today.year)
print('month :',today.month)
print('day :',today.day)


```
