# copy and view 



```python
# ex on view or  shallow copy 

arr=np.arange(10)
arr2=arr[:4]

print(arr)
print(arr2)
arr2[:]=99

print(arr)
print(arr2)

#change value of  arr2  afeect on arr1 becaise  arr2 is view of arr1
# tow refance  arr1,arr2 refer to same bloc in memory
```

```python
# ex on view or  deep copy 

arr=np.arange(10)
arr2=arr[:4].copy()

arr2
print(arr)
print(arr2)
arr2[:]=99

print(arr)
print(arr2)

#change value of  arr2  not afect on arr1 
```
